```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint
```

```
from google.colab import drive # uploading data
drive.mount('/content/gdrive')
```

```
admission_dataset = pd.read_csv('/content/gdrive/My Drive/Admission_Predict_Ver1.1.csv')
admission_dataset = admission_dataset.drop('Serial No.', axis=1)
```

```
predict_y = admission_dataset['Chance of Admit ']
predict_x = admission_dataset.drop('Chance of Admit ', axis=1)
```

```
X_train, X_test, y_train, y_test = train_test_split(predict_x, predict_y, test_size=0.2, random_state=42, shuffle=True)
```

```
model = Sequential()
model.add(Dense(200, activation='relu', kernel_initializer='random_normal', input_dim=X_test.shape[1]))
model.add(Dense(400, activation='relu', kernel_initializer='random_normal'))
model.add(Dense(4, activation='relu', kernel_initializer='random_normal'))
model.add(Dense(1, activation='linear', kernel_initializer='random_normal'))
opt = Adam(lr=1e-3, decay=1e-3 / 500)
model.compile(optimizer = opt, loss = 'mse', metrics= ['mse'])
```

```

checkpoint = ModelCheckpoint('/content/gdrive/My Drive/Colab Notebooks/test-least-loss4.h5', 
                             verbose=1, monitor='val_loss',save_best_only=True, mode='auto')
model.fit(X_train, y_train, batch_size=5, epochs=500, validation_data = (X_test, y_test), 
              callbacks = [checkpoint], verbose=1)
```
